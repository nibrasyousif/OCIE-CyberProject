CODE Cycle4/ Hybrid :(SecureBERT+LLMs)---LLMs model = genai.GenerativeModel('gemini-flash-latest')

```python
import spacy
import json
import time
import google.generativeai as genai

# 1. Setup API Key (Do not share this key on GitHub)
API_KEY = "your API Key "
genai.configure(api_key=API_KEY)

# Using the dynamic flash model for efficient and free access
llm_model = genai.GenerativeModel('gemini-flash-latest')

# 2. Load the trained Transformer model (SecureBERT)
print(" Loading trained SecureBERT model...")
nlp = spacy.load("/content/drive/MyDrive/OCIE_CyberProject1/output-Cycle3/model-best")

def extract_threat_actor_via_llm(text):
    short_text = text[:2500]
    prompt = f"""
    Analyze the following cyber threat report text and extract ONLY the exact name of the 'THREAT_ACTOR' group.
    Examples: APT29, LockBit, CyberAv3ngers, Scattered Spider, BlackTech, Snatch, Volt Typhoon, Phobos.
    If there is no specific threat actor group mentioned, return exactly the word: NONE.
    Do not use markdown, asterisks, or full sentences. Just the name.

    Text: "{short_text}"
    """
    try:
        # 5-second delay to prevent API quota limits
        time.sleep(5)
        response = llm_model.generate_content(prompt)
        result = response.text.strip().replace("**", "").replace("*", "")
        if "none" not in result.lower() and result != "":
            return result
        return None
    except Exception as e:
        print(f" 😢 [ Error connecting to LLM model: {e}]")
        return None

def evaluate_hybrid_pipeline(test_file_path):
    with open(test_file_path, 'r', encoding='utf-8') as f:
        data = json.load(f)

    total_found = 0
    resolved_by_llm = 0

    print("🧨\n Starting text analysis using the Final Hybrid Framework (Transformers + LLM)...\n")

    for entry in data:
        text = entry.get('text') or entry.get('data', {}).get('text', '')
        if not text: continue

        short_text_for_bert = text[:2000]

        # Phase 1: Transformers (SecureBERT) Extraction
        doc = nlp(short_text_for_bert)
        extracted_entities = {ent.label_: ent.text for ent in doc.ents}
        threat_actor = extracted_entities.get("THREAT_ACTOR")

        # False Positives Filtering
        is_false_positive = False
        if threat_actor:
            ta_lower = threat_actor.lower()
            if "cve" in ta_lower or "ransomware" in ta_lower or "stop" in ta_lower:
                is_false_positive = True

        # Phase 2: LLM Intervention for complex/unseen entities
        llm_result = None
        if not threat_actor or is_false_positive:
            llm_result = extract_threat_actor_via_llm(text)
            if llm_result:
                threat_actor = llm_result
                resolved_by_llm += 1
                is_false_positive = False

        # Output formatting
        if threat_actor and not is_false_positive:
            total_found += 1
            print(f"😊 Threat actor successfully detected: [{threat_actor}]")
        else:
            print("😒 No threat actor found in this report.")

    print("-" * 50)
    print(f" ✌️Total threat actors detected: {total_found}")
    print(f"👍 Number of complex entities successfully resolved by LLM: {resolved_by_llm}")
    print(" 👌Cycle 4 completed successfully!")

# Correct path to the test file
test_json_path = '/content/drive/MyDrive/OCIE_CyberProject1/test_cyber.json'
evaluate_hybrid_pipeline(test_json_path)
```
